This lab is about analysing the data that a large number of people colected on six species of birds.  Annas Hummingbird, Black Swift, Brambling, Harlequin Duck, Lazuli Bunting, Snowy Owl are analysed with respect to what time of year they show up, the size of their group and their location in King County.

```python
hummingbirdfile = open("AnnasHummingbird.csv", "r", errors = "ignore")
blackSwiftfile = open("BlackSwift.csv", "r", errors = "ignore")
bramblingfile = open("Brambling.csv", "r", errors = "ignore")
harlequinDuckfile = open("HarlequinDuck.csv", "r", errors = "ignore")
lazuliBuntingfile = open("LazuliBunting.csv", "r", errors = "ignore")
snowyOwlfile = open("SnowyOwl.csv", "r", errors = "ignore")
def listFromFile(file):
    list1 = []
    for i in file:
        list1.append(i)
    del list1[0]
    for j in range(len(list1)):
        list1[j]=list1[j].split(",")
    return list1
hummingbird = listFromFile(hummingbirdfile)
blackSwift = listFromFile(blackSwiftfile)
brambling = listFromFile(bramblingfile)
harlequinDuck = listFromFile(harlequinDuckfile)
lazuliBunting = listFromFile(lazuliBuntingfile)
snowyOwl = listFromFile(snowyOwlfile)
```

<h2>Location

```python
def long(bird):
    long = []
    for row in bird:
        long.append(row[22])
    return long
def lat(bird):
    lat = []
    for row in bird:
        lat.append(row[23])
    return lat
hummingbirdLong = long(hummingbird)
hummingbirdLat = lat(hummingbird)
blackSwiftLong = long(blackSwift)
blackSwiftLat = lat(blackSwift)
bramblingLong = long(brambling)
bramblingLat = lat(brambling)
harlequinDuckLong = long(harlequinDuck)
harlequinDuckLat = lat(harlequinDuck)
lazuliBuntingLong = long(lazuliBunting)
lazuliBuntingLat = lat(lazuliBunting)
snowyOwlLong = long(snowyOwl)
snowyOwlLat = lat(snowyOwl)
```

```python
import matplotlib.pyplot as plt
% matplotlib inline
plt.scatter(hummingbirdLat,   hummingbirdLong,   color='red',     s=0.05)
plt.scatter(blackSwiftLat,    blackSwiftLong,    color='orange',  s=4)
plt.scatter(bramblingLat,     bramblingLong,     color='yellow',  s=4)
plt.scatter(harlequinDuckLat, harlequinDuckLong, color='green',   s=4)
plt.scatter(lazuliBuntingLat, lazuliBuntingLong, color='blue',    s=4)
plt.scatter(snowyOwlLat,      snowyOwlLong,      color='purple',  s=4)
plt.title("Fig. 1: Location of Observation")
print("Hummingbird    = Red")
print("Black Swift    = Orange")
print("Brambling      = Yellow")
print("Harlequin Duck = Green")
print("Lazuli Bunting = Blue")
print("Snowy Owl      = Purple")
```

The Hummingbirds seem to be following a pattern of showing up wherever there are observers.  The map above seems to be if only looking at the red humming bird points to practically be a population density map. Which probably means that there were far fewer people going out of their way to try and find hummingbirds than the ones who just spotted one in their back yard and reported it. However with the other birds there is mostly not mush of a pattern at all except the duck which forms a bit of a line, probably due to a water sourse and/or their migration pattern.

<h2>Time

```python
def months(bird):
    birdMonths = []
    for i in range(12):
        birdMonths.append(0)
    for row in bird:
        if row[24][1]== '/':
            row[24]= '0'+row[24]
        for j in range(12):
            if float(row[24][0:2]) == j+1:
                if row[7]== 'X':
                    birdMonths[j]+=1
                else:
                    birdMonths[j]+=float(row[7])
    return birdMonths
def season(bird):
    birdSeason = []
    for i in range(4):
        birdSeason.append(0)
    for row in bird:
        if row[24][1]== '/':
            row[24]= '0'+row[24]
        for j in range(4):
            if int(row[24][0:2]) < 4:
                if row[7]== 'X':
                    birdSeason[0]+=1
                else:
                    birdSeason[0]+=float(row[7])
            elif int(row[24][0:2]) < 7:
                if row[7]== 'X':
                    birdSeason[1]+=1
                else:
                    birdSeason[1]+=float(row[7])
            elif int(row[24][0:2]) < 10:
                if row[7]== 'X':
                    birdSeason[2]+=1
                else:
                    birdSeason[2]+=float(row[7])
            else:
                if row[7]== 'X':
                    birdSeason[3]+=1
                else:
                    birdSeason[3]+=float(row[7])
    return birdSeason
hummingbirdMonths   = months(hummingbird)
hummingbirdSeason   = season(hummingbird)
blackSwiftMonths    = months(blackSwift)
blackSwiftSeason    = season(blackSwift)
bramblingMonths     = months(brambling)
bramblingSeason     = season(brambling)
harlequinDuckMonths = months(harlequinDuck)
harlequinDuckSeason = season(harlequinDuck)
lazuliBuntingMonths = months(lazuliBunting)
lazuliBuntingSeason = season(lazuliBunting)
snowyOwlMonths      = months(snowyOwl)
snowyOwlSeason      = season(snowyOwl)
```

```python
def month (x):
    monthlist = []
    for i in range(12):
        monthlist.append(i+x+1)
    return monthlist
def dev (bird):
    list1=[]
    for i in range(len(bird)):
        list1.append(bird[i]/max(bird)*100)
    return list1
plt.bar(month(-2*.16),dev(hummingbirdMonths),width=0.16,color='red')
plt.bar(month(-1*.16),dev(blackSwiftMonths),width=0.16,color='orange')
plt.bar(month(0*.16),dev(bramblingMonths),width=0.16,color='yellow')
plt.bar(month(1*.16),dev(harlequinDuckMonths),width=0.16,color='green')
plt.bar(month(2*.16),dev(lazuliBuntingMonths),width=0.16,color='blue')
plt.bar(month(3*.16),dev(snowyOwlMonths),width=0.16,color='purple')
plt.xlabel('Month')
plt.ylabel('Sightings (% of max)')
plt.title('Fig. 2: A Comparison of the Month Birds are Observed')
print("Hummingbird    = Red")
print("Black Swift    = Orange")
print("Brambling      = Yellow")
print("Harlequin Duck = Green")
print("Lazuli Bunting = Blue")
print("Snowy Owl      = Purple")
```

The diagram above clearly shows that in different months different birds dissapear due to the seasonal temperature. it shows that in specific months some birds show up while others avoid the area.  

```python
def seasons (x):
    seasonlist = []
    for i in range(4):
        seasonlist.append(i+x+1)
    return seasonlist
plt.bar(seasons(-2* .16), dev(hummingbirdSeason),   width=0.16, color='red')
plt.bar(seasons(-1* .16), dev(blackSwiftSeason),    width=0.16, color='orange')
plt.bar(seasons(0 * .16), dev(bramblingSeason),     width=0.16, color='yellow')
plt.bar(seasons(1 * .16), dev(harlequinDuckSeason), width=0.16, color='green')
plt.bar(seasons(2 * .16), dev(lazuliBuntingSeason), width=0.16, color='blue')
plt.bar(seasons(3 * .16), dev(snowyOwlSeason),      width=0.16, color='purple')
plt.xlabel('Winter, Spring, Summer, Fall')
plt.ylabel('Sightings (% of max)')
plt.title('Fig. 3: A Comparison of the Season Birds are Observed')
```

Figure 3 combined with Figure 2 shows the strong seasonal preferace of the birds.  You can see in the two charts that the Snowy Owl and the Harlequin Duck appeared in the winter quite a bit while the Black Swift and the Lazuli Bunting preferred the summer.

<h2>Observation Count

```python
def appear(bird):
    list2 = []
    for row in bird:
        if row[7]=='X':
            list2.append(1)
        elif int(row[7]) < 25:
            list2.append(int(row[7]))
    return list2
plt.hist(appear(hummingbird), color = 'red')
plt.title('Fig. 4: Birds that Appear at the Same Time')
print("Hummingbird    = Red")
print("Black Swift    = Orange")
print("Brambling      = Yellow")
print("Harlequin Duck = Green")
print("Lazuli Bunting = Blue")
print("Snowy Owl      = Purple")
```

```python
plt.hist(appear(blackSwift), color = 'orange')
plt.title('Fig. 5: Birds that Appear at the Same Time')
```

```python
plt.hist(appear(harlequinDuck), alpha = 0.5, color='green')
plt.hist(appear(lazuliBunting), alpha = 0.5, color='blue')
plt.title('Fig. 6: Birds that Appear at the Same Time')
```

```python
plt.hist(appear(snowyOwl), alpha = 0.5, color='purple')
plt.hist(appear(brambling), alpha = 0.5, color='yellow')
plt.title('Fig. 7: Birds that Appear at the Same Time')
```

Figures 4-7 show how many birds people counted in their observations.  I ommitted the data on the charts for when people counted more than 25 due to those values being too small to really show up on my histograms.  People mostly only saw one or two birds at a time for all of the spicies with rare sightings of more and getting even more rare as the number of birds goes up.

It appears that the birds mostly show up in groups of ones and twos when they are counted, The location of the Observation seems to be mostly random except for the Duck which likes water, and The birds like to move away when they do not like the temperature.

```python
import statistics as st
import scipy.stats as sci
import matplotlib.pyplot as plt
% matplotlib inline

victimsFile = open("KilledByPolice.csv","r", errors = "ignore")
cityRatesFile = open("PoliceViolence_city.csv","r", errors = "ignore")

def listFromFile(file):
    list1 = []
    for i in file:
        list1.append(i)
    del list1[0]
    for j in range(len(list1)):
        list1[j]=list1[j].split(",")
    return list1
def columnToList(list1, column):
    list2=[]
    for row in list1:
        list2.append(float(row[column]))
    return list2

victims = listFromFile(victimsFile)
cityRates = listFromFile(cityRatesFile)

policeHomicideRate=columnToList(cityRates, 20)
violentCrimeRate=columnToList(cityRates, 18)
plt.scatter(violentCrimeRate, policeHomicideRate)
plt.title("People Killed by Police Vs. Violent Crime")
plt.xlabel("Crime Rate")
plt.ylabel("Police Homicide Rate")

regression = sci.linregress(violentCrimeRate, policeHomicideRate)
slope = regression[0]
y_int = regression[1]
plt.plot([0, 25], [y_int, (slope * 25) + y_int])
print("y =", slope, "* x +", y_int)
```

```python
totalPop = columnToList(cityRates, 5)
blackPop = columnToList(cityRates, 6)
percentBlack = []
for i in range(len(blackPop)):
    percentBlack.append(blackPop[i]/totalPop[i])
policeBlackHomicideRate = columnToList(cityRates, 21)
print(policeBlackHomicideRate[0])

plt.scatter(percentBlack, policeHomicideRate, color='red')
plt.scatter(percentBlack, policeBlackHomicideRate)

plt.title("Percent Black Vs. Police Black Homicide Rate")
plt.xlabel("Percent Black")
plt.ylabel("Police Black Homicide Rate (blue)")

regression = sci.linregress(percentBlack, policeBlackHomicideRate)
slope = regression[0]
y_int = regression[1]
plt.plot([0, 1], [y_int, (slope * 1) + y_int])
print("y =", slope, "* x +", y_int)
print("r =",regression[2])
```
